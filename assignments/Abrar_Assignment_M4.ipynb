2a. follow the lecture examples and create a function that will calculate the area of a rectangle.
Then call it and calculate the area of a 4x8 rectangle.

2b. Write another function that calculates the area of a square, in which you need to call the function
you just created in Q2a

```python
def area_rect(length: int, width: int) -> int:
    """
    This function calculates the area of a rectangle.
    """
    return length * width

print(area_rect(4, 8))

def area_square(side: int) -> int:
    """
    This function calculates the area of a square.
    """
    return area_rect(side, side)

print(area_square(4))
```

Write a Python program to construct the following step pattern with any number of rows. Call your
function to print a 3-row pattern. Call it again to print a 6-row pattern. Hint: to repeat a character,
you can use the * operator.

Expected Output:

1

22

333

1

22

333

4444

55555

666666

```python
def three_row_patterns(rows: int) -> None:
    """
    This function prints a pattern of three rows with asterisks.
    """
    for i in range(1, rows + 1):
        number_output = str(i) * i
        print(number_output)


print("3 Row Pattern:")
three_row_patterns(3)

print("------------------------")

# Call the function to print a 6-row pattern
print("6 Row Pattern:")
three_row_patterns(6)
```

Practice using arbitrary arguments. You are designing a simple report that will count the number of
kids and print out a message “There are <n> kids in total.”

4a. Create a function that will receive an arbitrary argument with a bunch of names and print out
that message with the right count.

4b. Call your function with these names in a tuple: Jame, Michael, Kevin, Kate, Matthew, Jacob

4c. Call your function with these names in a list: Abel, Beth, Charlie, Derrick, Paul


```python
# 4a
def count_kids(*kids):
    """
    This function counts the number of kids.
    """
    count_total = len(kids)
    return(f"There are {count_total} in total")

# 4b
names_tuple = ("Jame", "Michael", "Kevin", "Kate", "Matthew", "Jacob")
print(count_kids(*names_tuple))

# 4c
names_list = ["Abel", "Beth", "Charlie", "Derrick", "Paul"]
print(count_kids(*names_list))
```
