Q1. Plotting multiple variables.

1.1 Use numpy linspace function to create a numpy array with 60 numbers between 0 and 6, assign the array to variable x

1.2 Create numpy array y = x**2 + x +1, and plot a line using x and y

1.3 Create numpy array z = x**3 + x**2 + x +1, and plot a line using x and z

1.4 Plot two lines (x, y) and (x, z) in the same figure, the first line is a solid black line with
circle marker, while the second line is a dashed green line with diamond marker.
Generate the legend with the first line’s label set to ‘y’, and second line’s label set to ‘z’

1.5 Choose your favorite plot style from plt.style.available, and plot the above figure again.

1.6 Plot the above 3 plots in one panel with subplots.

```python
# Q1
import numpy as np
import matplotlib.pyplot as plt

# Q1.1
x = np.linspace(0, 10, 100)

# Q1.2
y = x**2 + x + 1
plt.figure(figsize=(6, 6))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Q1.2 Plot: y = x^2 + x + 1')
plt.grid(True)
plt.show()

# Q1.3
z = x**3 + x**2 + x + 1
plt.figure(figsize=(6, 6))
plt.plot(x, z)
plt.xlabel('x')
plt.ylabel('z')
plt.title('Q1.3 Plot: z = x^3 + x^2 + x + 1')
plt.grid(True)
plt.show()

# Q1.4
plt.figure(figsize=(6, 6))
plt.plot(x, y, 'k-o', label='y', markevery=10)  # Black line with circles
plt.plot(x, z, 'g--D', label='z', markevery=10)  # Green dashed line with diamonds
plt.xlabel('x')
plt.ylabel('Value')
plt.title('Q1.4 Plot: Comparison of y and z')
plt.legend()
plt.grid(True)
plt.show()

# Q1.5
print("Styles:", plt.style.available)
plt.style.use('Solarize_Light2') # Was the first one, has a cool name too

plt.figure(figsize=(6, 6))
plt.plot(x, y, 'k-o', label='y', markevery=10)
plt.plot(x, z, 'g--D', label='z', markevery=10)
plt.xlabel('x')
plt.ylabel('Value')
plt.title('Q1.5 Plot: Comparison of y and z (Using Solarize Light2 Style)')
plt.legend()
plt.show()

# Q1.6
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# First subplot: y only
axes[0].plot(x, y)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Q1.6 Plot: y = x^2 + x + 1')
axes[0].grid(True)

# Second subplot: z only
axes[1].plot(x, z)
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
axes[1].set_title('Q1.6 Plot: z = x^3 + x^2 + x + 1')
axes[1].grid(True)

# Third subplot: both y and z
axes[2].plot(x, y, 'k-o', label='y', markevery=10)
axes[2].plot(x, z, 'g--D', label='z', markevery=10)
axes[2].set_xlabel('x')
axes[2].set_ylabel('Value')
axes[2].set_title('Q1.6 Plot: Comparison of y and z')
axes[2].legend()
axes[2].grid(True)

plt.tight_layout()
plt.show()
```

Q2. Generating random samples and plotting histograms.

Q2.1 Use numpy random.normal function to create a numpy array of 2000 normally distributed
random values, this time specify the mean and standard deviation of the distribution as 4
and 3. Assign the array to variable n, plot the histogram of numpy array n. Pay attention
to your bin size.

Q2.2 Import the seaborn package, and call set function, plot the histogram of numpy array n
again.

Q2.3 It’s impossible to go through every available option in any plot function. Research on the
seaborn’s kdeplot function, and then plot the KDE of the numpy array n.

Q2.4 Do a box plot of n, and another box plot going horizontally. Place these two plots as
subplots in one figure.


```python
# Q2
import seaborn as sns

# Set seed for reproducibility
np.random.seed(42)

# Q2.1
n = np.random.normal(loc=4, scale=3, size=2000)

plt.figure(figsize=(6, 6))
plt.hist(n, bins=30, edgecolor='black', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Q2.1 Plot: Histogram of Normally Distributed Random Values')
plt.grid(True, alpha=0.3)
plt.show()

print(f"Mean: {np.mean(n):.2f}")
print(f"Standard deviation: {np.std(n):.2f}")

# Q2.2 
sns.set()
plt.figure(figsize=(6, 6))
plt.hist(n, bins=30, edgecolor='black', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Q2.2 Plot: Histogram with Seaborn Styling')
plt.grid(True, alpha=0.3)
plt.show()

# Q2.3
# KDE plots a smooth curve showing where most data points cluster together
# Higher curve means more data points aer in that range.
plt.figure(figsize=(6, 6))
sns.kdeplot(data=n, fill=True, color='blue', alpha=0.5)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Q2.3 Plot: KDE of Normally Distributed Data')
plt.grid(True, alpha=0.3)
plt.show()

# Q2.4 
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Vertical box plot
axes[0].boxplot(n, vert=True)
axes[0].set_ylabel('Value')
axes[0].set_title('Q2.4 Plot: Vertical Box Plot')
axes[0].grid(True, alpha=0.3)

# Horizontal box plot
axes[1].boxplot(n, vert=False)
axes[1].set_xlabel('Value')
axes[1].set_title('Q2.4 Plot: Horizontal Box Plot')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
