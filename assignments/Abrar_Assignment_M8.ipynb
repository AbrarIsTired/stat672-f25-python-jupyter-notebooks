Question #1

Create a Pandas DataFrame from the following dictionary data and the index labels

exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],

'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],

'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],

'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'] ,

then do the following: 

•	  Print the first 3 rows of the dataFrame

•	  Print the 'name', 'score' columns

•	  Fill the missing score with number 0.0


```python
# Question #1
import pandas as pd
import numpy as np

# Dictionary Setup
exam_data = {
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
}

# Index 
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

# DataFrame
question_1_df = pd.DataFrame(exam_data, index=labels)

print("DataFrame:")
print(question_1_df)
print("\n" + "#"*60 + "\n") # Added seperator for clarity

# Print the first 3 rows
print("First Three Rows:")
print(question_1_df.head(3))
print("\n" + "#"*60 + "\n")

# Print 'name' and 'score' columns
print("Name and Score Cols:")
print(question_1_df[['name', 'score']])
print("\n" + "#"*60 + "\n")

# Fill missing scores with 0.0
question_1_df['score'].fillna(0.0, inplace=True)

# Print the updated DataFrame
print(question_1_df)
```

Question #2

For the dataframe you created in Q1, we want to add the timestamps of when the students took the exam.  Assume the first student started at 8am, on Oct 23, 2023, and each successive student took it exactly one hour after the other. 

•	Create the appropriate DatetimeIndex series, then apply it as index to the dataframe.

•	Next, add a column called “ExamTime” which will have the timestamp values as you just created.

•	Suppose scores will be published 7 days after the exam date.  Add a column “ScoreDate” which will have the date of the score published. Hint: use the Timedelta() function.

•	If your score_date thus created has the h:m:s info in it, format it to show just the date. Hint: use .dt.date for this part.


```python
# Question 2

# Create DatetimeIndex 
start_time = pd.Timestamp('2023-10-23 08:00:00')
datetime_index = pd.date_range(start=start_time, periods=len(question_1_df), freq='h')

# Putting DatetimeIndex to the dataframe
question_1_df.index = datetime_index
print(question_1_df)
print("\n" + "#"*60 + "\n")

# Add ExamTime column with timestamp values
question_1_df['ExamTime'] = question_1_df.index
print(question_1_df)
print("\n" + "#"*60 + "\n")

# Add 'ScoreDate' column
question_1_df['ScoreDate'] = question_1_df['ExamTime'] + pd.Timedelta(days=7)
print(question_1_df)
print("\n" + "#"*60 + "\n")

# Format ScoreDate
question_1_df['ScoreDate'] = question_1_df['ScoreDate'].dt.date

print("Final DataFrame with 'ScoreDate' formatted as date only:")
print(question_1_df)
```
