Question #3
To continue on Assignment 5 Question 3, where we used string functions to find the state(s) with the longest name. There are actually two states of the same name length. Add a for loop or a while loop to print both states.

```python
#Question #3

# From Assignment #5 Question 3
states = ['Alabama','Alaska','Arizona','Arkansas','California','Colorado',
'Connecticut','Delaware','Florida','Georgia','Hawaii','Idaho',
'Illinois','Indiana','Iowa','Kansas','Kentucky','Louisiana',
'Maine','Maryland','Massachusetts','Michigan','Minnesota',
'Mississippi','Missouri','Montana','Nebraska','Nevada',
'New Hampshire','New Jersey','New Mexico','New York',
'North Carolina','North Dakota','Ohio','Oklahoma','Oregon',
'Pennsylvania','Rhode Island','South Carolina','South Dakota',
'Tennessee','Texas','Utah','Vermont','Virginia','Washington',
'West Virginia','Wisconsin','Wyoming']

max_length = max(len(state) for state in states) 
print(f"States with the longest name length ({max_length} characters):")
for state in states:
    if len(state) == max_length:
        print(f"  {state}")
```

Question #4
Q4. Below are two lists:

List1=[1,2,3,4]
List2=[5,6,7,8]

4a. Use list comprehension to create a new list from List1, where odd numbers will be squared and
even numbers will be doubled. Result should be [1, 4, 9, 8].

4b. Use list comprehension to create a new list with element-wise product of List1 and List2. Result
should be [5, 12, 21, 32]. Hint: product=x*y, and you need to use nested for loops.


```python
List1 = [1, 2, 3, 4]
List2 = [5, 6, 7, 8]

# Q4A
result_a = [x**2 if x % 2 == 1 else x*2 for x in List1]
print(f"{result_a}")

# Q4B
result_b = [x*y for x, y in zip(List1, List2)]
print(f"{result_b}")
```
