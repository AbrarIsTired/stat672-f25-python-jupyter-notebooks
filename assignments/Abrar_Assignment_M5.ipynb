Question #1:
What are the 3 common ways of string formatting? How do they compare with each other?

Answer: 
The old way is the % operator, which is similar to C's printf style. It's almost like an f-string, but you have to specify the type of variable you are inserting. You would use %s for string, %d for integer. Then you have .format which works almost like a function with parameters. You put {} in the string where you want to insert a variable, and then call .format(variable) after the string. Finally  f-strings where you put an f before the string and then you can insert variables directly into the string using {}. They are also the easiest to read and write.

% Operator:

name = "Abrar"

age = 21

print("Hello! I'm %s and I am %i years old." % (name, age))

.format:

print("Hello! I'm {} and I am {} years old.".format("Abrar", 21))

f-string:

name = "Abrar"

age = 21

print(f"Hello! I'm {name} and I am {age} years old.")

```python
# Question #2

# Timer Code from slide 11 of M5a
qty =list(range(0, 100000))
qty_set=set(qty)

import time

start = time.perf_counter()
len(qty)
end = time.perf_counter()
tlist = end - start
print("time elapse using list: ", tlist)

start = time.perf_counter()
len(qty_set)
end = time.perf_counter()
tset = end - start
print("time elapse using set: ", tset)

print("time using list is %0.2f times of using set" %(tlist/tset))


# Create a between 2A and original code
print("-" * 40)
print("-" * 40)

#2A
def timer_func(data, description):
    """
    Times how long an operation takes. Uses len() as the operation to test.

    Aruguments:
    data: the data structure to test (e.g., list, set, dict)
    description: a string description of the data structure

    Returns:
    elapsed: the time taken to perform the operation of len(data)
    """
    start = time.perf_counter()
    len(data)
    end = time.perf_counter()
    elapsed = end - start
    print(f"Time for {description}: {elapsed:.8f} seconds")
    return elapsed

#2B

# Make a list of numbers
my_list = list(range(100000))

# Make a set from the same numbers
my_set = set(my_list)

# Make a dictionary with numbers as keys and values
my_dict = {i: i*2 for i in range(100000)}

# Timer function on each data type
time_list = timer_func(my_list, " My list")
time_set = timer_func(my_set, "My set based on list") 
time_dict = timer_func(my_dict, "My dictionary")

# Results
print(f"List time: {time_list:.8f} seconds")
print(f"Set time: {time_set:.8f} seconds") 
print(f"Dict time: {time_dict:.8f} seconds")
```
