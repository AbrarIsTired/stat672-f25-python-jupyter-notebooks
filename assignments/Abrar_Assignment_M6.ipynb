Q1. Use your own words to explain why a NumPy array is more efficient than a list or dictionary

NumPy Arrays are faster as they store all elements of the same data type in contiguous memory locations. This allows you to retrieve and manipulate data more quickly than lists or dictionaries, which can store elements of different data types but may not be stored in contiguous memory locations. There is also an underlying C Compiler which is a low level language and is known to be very fast. You also dont need to use loops as it"ll "crunch" the numbers in the array opposed to looping through each element by element.

Q2. On NumPy. This question has the following 8 steps:

1) Create a Pandas Series s1 with values: "New Jersey", "Delaware", "Pennsylvania", "Maryland",
and index "a", "b", "c", "d"

2) Create a Pandas Series s2 with values: "California", " Florida", " Illinois", "Connecticut",
and index "e", "f", "g", "h"

3) Concatenate Series s1 and s2 together, and assign it to s3

4) Create a Pandas Series s4 with values:
"Trenton", "Dover", "Harrisburg", "Annapolis", "Sacramento", "Tallahassee", "Springfield", "Hartford",
and indexes "a", "b", "c", "d", "e", "f", "g", "h"

5) Create a Pandas Series s5 with values:
581.12, 70.39, 724.94, 378.28, 2600.0, 926.82, 791.61, 263.38
and index "a", "b", "c", "d", "e", "f", "g", "h"

6) Create a dataFrame d1 with s3, s4, s5, and column names: "state", "capital", "GDP

```python
import pandas as pd
import numpy as np

# Q2.1
s1 = pd.Series(["New Jersey", "Delaware", "Pennsylvania", "Maryland"], index=["a", "b", "c", "d"])


# Q2.2
s2 = pd.Series(["California", "Florida", "Illinois", "Connecticut"], index=["e", "f", "g", "h"])


# Q2.3
s3 = pd.concat([s1, s2])


# Q2.4
s4 = pd.Series(["Trenton", "Dover", "Harrisburg", "Annapolis", "Sacramento", "Tallahassee", "Springfield", "Hartford"], index=["a", "b", "c", "d", "e", "f", "g", "h"])


# Q2.5
s5 = pd.Series([581.12, 70.39, 724.94, 378.28, 2600.0, 926.82, 791.61, 263.38],index=["a", "b", "c", "d", "e", "f", "g", "h"])


# Q2.6
d1 = pd.DataFrame({"State": s3, "Capital": s4, "GDP": s5})
```

Q3. On NumPy and random numbers.

3a. Create a random sampe of 100 normal random variables, and put them into an 10x10 array. For
the array, calculate the column wise means and save to an array. Hint: the argument axis=0 will
return column wise means for the mean() methods.

3b. similarly, calculate the column wise minimums. Then calculate the column wise maximums.
Save to respective arrays.

3c. The mean should be >= min, and <= max. Use comparison operators to find out whether this
holds for the results in 3a and 3b.


```python
import numpy as np
# import random : I forgot Numpy has its own random module

# Q3.A 
np.random.seed(42) 
random_array = np.random.randn(10, 10)
print("10x10 Array")
print(random_array)


column_means = random_array.mean(axis=0)
print("Col Means:")
print(column_means)


# Q3.B 
column_mins = random_array.min(axis=0)
print("Col Mins:")
print(column_mins)


column_maxs = random_array.max(axis=0)
print("Col Max:")
print(column_maxs)


# Q3.C 
mean_greater_equal_min = column_means >= column_mins
mean_less_equal_max = column_means <= column_maxs


print("mean >= min for each col?")
print(mean_greater_equal_min)
print(f"All True? {mean_greater_equal_min.all()}")


print("mean <= max for each col?")
print(mean_less_equal_max)
print(f"All True? {mean_less_equal_max.all()}")


print("Do both conditions hold for all cols?")
print(f"{(mean_greater_equal_min & mean_less_equal_max).all()}")
```
