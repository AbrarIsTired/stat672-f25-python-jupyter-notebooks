Question #1
Based on the lectures of module 3, do you think tuples can serve as arguments to a function?
Construct a small Python program to show your answer. Be sure to document what you are trying to
do and use a comment to answer this question.

```python
#Question #1
def tuple_as_argument(my_tuple):
    """
    Using Tuples as an argument in function
    
    Arguments: my_tuple: A tuple passed as an argument
    
    Returns: Returns the Tuple, Length and Type
    """
    return(f"Received tuple: {my_tuple}, Type: {type(my_tuple)}, Length: {len(my_tuple)}")

def multi_tuple_arguments(*my_tuple):
    """
    Unpacking tuple arguments using *args (One or more arguments packed together)
    
    Arguments: *my_tuple: Variable number of arguments
    
    Returns: str: Information about the tuple
    """
    return(f"Unpacked arguments: {my_tuple}, Length: {len(my_tuple)}")

# Pass tuple directly as a single argument
tuple_one = (1, 2, 3, 'abrar', 'test', 672)
result1 = tuple_as_argument(tuple_one)
print(result1)

# Unpack tuple into multiple arguments using *
tuple_two = multi_tuple_arguments(*tuple_one)
print(tuple_two)
```

Modify the code in the lecture to draw a shape with 24 equal sides, and the color of the sides should
cycle through 5 colors of your choice. Use comments to explain what your code line was supposed
to do. 

If your computer doesn’t support the mobilechelonian package, there are two workarounds- to

```python
#Question #2
import turtle

# Hello, Prof! I had to restart the Python Kernel due to it crashing with the Turtle every other run, the window would pop up
# but then would crash. I would re-run, press restart Python Kernel and it would draw. Just wanted to give a heads up

def draw_24_sided_polygon():
    """
    Drawing a 24-sided polygon with turtle
    Arguments: None
    
    Returns: A Turtle Window with what is a 24-sided polygon
    """

    # Create a turtle
    t = turtle.Turtle()
    t.speed(6)
        
    # Define 5 colors to cycle through. Easier to have it cycle through a list then manually draw it all out with pencolor every single time
    colors = ["red", "blue", "green", "orange", "purple"]
        
    # Finding the angle for a 24-sided polygon for each turn
    angle = 360 / 24  # This equals 15 degrees per turn
    side_length = 25  # Something small to fit within the screen, my laptop is kind of tiny
    count = 0 # Storing index to use to cycle through the color list
    # Draw all 24 sides using a simple loop
    for i in range(24):
        current_color = colors[count]
        t.pencolor(current_color)
            
        # Draw one side: move forward, then turn right
        t.forward(side_length)
        t.right(angle)
        count = count + 1
        # Reset the index back to 0. There are probably better ways to do this but this is the one that came to mind
        if(count >= 5):
            count = 0
    # Keep the turtle window open. I am using a Jupyter Notebook on VSC
    turtle.done()

draw_24_sided_polygon()    
```
